```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

market = pd.read_csv('/content/Products.csv')

market['Total_Sales'] = market['Product_Price']

market_sales = market.groupby('Market')['Total_Sales'].sum().reset_index()
market_sales['Total_Sales'] = market_sales['Total_Sales'].round(-3).astype(int)

total_sales_per_market = market.groupby('Market')['Product_Price'].sum()

sales_per_category = market.groupby(['Market', 'Category_Name'])['Product_Price'].sum().reset_index()

merged_df = pd.merge(sales_per_category, total_sales_per_market, on='Market', suffixes=('_Category', '_Total'))

merged_df['Percentage'] = (merged_df['Product_Price_Category'] / merged_df['Product_Price_Total']) * 100

merged_df[['Market', 'Category_Name', 'Percentage']].head()
```

```python
fig, axis = plt.subplots()

colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
bars = market_sales.plot.bar(x='Market', y='Total_Sales', ax=axis, color=colors, legend=False)

axis.set_xlabel('Market')

axis.set_ylabel('Total Sales')

axis.set_title('Total Sales by Market')
for bar in bars.patches:
    axis.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height()}',
        ha='center',
        va='bottom'
    )

plt.show()
```

```python
fig = px.bar(market_sales,
             x='Market',
             y='Total_Sales',
             title='Total Sales by Market',
             labels={'Market': 'Market', 'Total_Sales': 'Total Sales'},
             text='Total_Sales',
             color='Market')

fig.update_layout(
    xaxis_title='Market',
    yaxis_title='Total Sales',
    uniformtext_minsize=8,
    uniformtext_mode='show',
)

fig.show()
```

```python
selected_market = 'LATAM'
filtered_df = merged_df[merged_df['Market'] == selected_market]

small_categories = filtered_df[filtered_df['Percentage'] < 1]

other_percentage = small_categories['Percentage'].sum()

remaining_categories = filtered_df[filtered_df['Percentage'] >= 2]

other_row = pd.DataFrame({'Market': [selected_market], 'Category_Name': ['Other'], 'Percentage': [other_percentage]})
final_df = pd.concat([remaining_categories, other_row], ignore_index=True)

fig = px.pie(final_df, values='Percentage', names='Category_Name', title=f'Percentage of Sales by Category in {selected_market}',
             custom_data=['Percentage'])

fig.update_traces(hovertemplate='%{label}: %{percent}<br>Percentage: %{customdata[0]:.2f}%')

fig.show()
```
