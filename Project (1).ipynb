QUERY 1:

USE Spr_2024;
WITH SegmentCounts AS (
   SELECT Customer_Segment, COUNT(*) AS Segment_Count
   FROM CustomerData
   GROUP BY Customer_Segment)
SELECT Market, CD.Customer_Segment,
   (SELECT COUNT(DISTINCT Market) FROM ShippingData) AS Market_Count,
   MAX(SC.Segment_Count) AS Customer_Segment_Count,
   Customer_Country
FROM ShippingData AS SD
JOIN MainTable AS MT ON MT.Order_Id = SD.Order_Id
JOIN CustomerData AS CD ON CD.Customer_Id = MT.Customer_Id
JOIN SegmentCounts AS SC ON CD.Customer_Segment = SC.Customer_Segment
GROUP BY Market, CD.Customer_Segment, Customer_Country;

```python
merged_data = pd.merge(customer_data, order_data, on=['Market', 'Customer_Segment'], how='inner')

merge_df= pd.DataFrame(merged_data)
merge_df.head(30)
```

```python
total_orders_sum = merged_data.total_orders.sum()
print(total_orders_sum)
```

#Top Orders by Market and Segment

```python
total_orders_by_market_segment = merged_data.groupby(['Market', 'Customer_Segment'])['total_orders'].sum().reset_index()
total_orders_by_market_segment = total_orders_by_market_segment.sort_values(by='total_orders', ascending=False)
total_orders_by_market_segment.head()
```

```python
top_markets = merged_data.groupby('Market')['total_orders'].sum().sort_values(ascending=False)
top_markets_df = pd.DataFrame(top_markets)
top_markets_df.head()
```

```python
top_markets_df['order_ratio'] = (top_markets_df['total_orders'] / total_orders_sum) *100
top_markets_df.head()
```
